```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temp_model(time, a, b, c):
    # Quadratic equation: t = at^2 + bt + c
    temperature = a * (time**2) + b * time + c
    return temperature

def read_coefficients_from_file(file_path):
    coefficient_sets = []
    with open(file_path, 'r') as file:
        for line in file:
            coefficients = list(map(float, line.split()))
            coefficient_sets.append(coefficients)
    return coefficient_sets

def main():
    file_path = 'coefficients.txt'
    coefficient_sets = read_coefficients_from_file(file_path)
    time_values = np.linspace(0, 10, 51)

    # Plotting each set of coefficients
    for coefficients in coefficient_sets:
        a, b, c = coefficients
        temperature_values = quadratic_temp_model(time_values, a, b, c)
        label = f"a={a}, b={b}, c={c}"
        plt.plot(time_values, temperature_values, label=label)

    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title("Multiple Sets Weather Modeling")
    plt.show()

if __name__ == "__main__":
    main()
```
